# Exploring Data Visualization Using Matplotlib

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./world-development-indicators/Indicators.csv')
data.shape
```

```python
# creating python lists for each individual country and world development indicator

countries = data['CountryName'].unique().tolist()
indicators = data['IndicatorName'].unique().tolist()
```

Looks like it has different indicators for different countries with the year and value of the indicator. 

We already saw how the USA's per-capita CO2 production related to other countries, let's see if we can find some more indicators in common between countries.  

To have some fun, we've picked countries randomly but then stored our random results so you can rerun it with the same answers.

```python
# Filter 1 - remember filters are the same as masks

# Picks years of choice
yearsFilter = [2010, 2011, 2012, 2013, 2014]
```

```python
# Filter 2 - the random.sample method returns a particular length list of items chosen from the sequence i.e. list, tuple, 
# string or set. Used for **random sampling** without replacement. Syntax: random.sample(sequence, k)

# Parameters:
# sequence: Can be a list, tuple, string, or set.
# k: An Integer value. it specifies the length of a sample.

# Note: When assigning random.sample to a variable, the returned sample stays with the variable until it is reassigned

# Below picks 2 countries randomly from our list 'countries'
countryFilter = random.sample(countries, 2)
```

```python
countryFilter
```

```python
# Filter 3

# Picks 1 Indicator randomly from our list 'indicators'
indicatorsFilter = random.sample(indicators, 1)
indicatorsFilter
```

# Problem:  We're missing data.
# Not all countries have all indicators for all years

To solve this, we'll need to find two countries and two indicators for which we have data over this time range.

```python
# by using the '&' operator, we ensure that filterMesh combines all of our filter criteria

# remember, the Pandas isin() method is used to filter data frames. isin() helps in selecting rows for which have a
# particular (or Multiple) value in a particular column -> can pass in a list if we want, as demonstrated for 'Year' below.

# the Pandas .loc() method is used to retrieve rows from a Data frame. DataFrame.loc[] method is a method that takes only
# index labels and returns rows or a Dataframe if the index label(s) exists in the caller Dataframe

filterMesh = (data['CountryName'] == countryFilter[0]) & (data['IndicatorName'].isin(indicatorsFilter)) & \
            (data['Year'].isin(yearsFilter))
country1_data = data.loc[filterMesh]

len(country1_data)
```

```python
country1_data
```

```python
# we do the same filter process for our second country. we need to ensure that it has the same amount of data as country one

filterMesh = (data['CountryName'] == countryFilter[1]) & (data['IndicatorName'].isin(indicatorsFilter)) & \
            (data['Year'].isin(yearsFilter))
country2_data = data.loc[filterMesh]

len(country2_data)
```

```python
country2_data
```

# So let's pick indicators and countries which have data over this time range

The code below will randomly pick countries and indicators until it finds two countries who have data for an indicator over this time frame. We used it to produce the fixed values you see later, feel free to play with this yourself!

```python
filteredData1 = []
filteredData2 = []
```

```python
'''
Plot: 
countryFilter:    pick two countries, 
indicatorsFilter: pick an indicator, 
yearsFilter: plot for years in yearsFilter
'''
# problem - not all countries have all indicators so if you go to visualize, it'll have missing data.
# randomly picking two indicators and countries, do these countries have valid data over those years?
# brings up the discussion of missing data/ missing fields until we find full data

# this can be dealt with by using a while loop as below

# the while loop is True until countries and indicators with 5 years of data are found - at which point it becomes
# False, and the loop ends
while(len(filteredData1) < len(yearsFilter)-1):
    # below randomly picks a new indicator and 2 new countries
    indicatorsFilter = random.sample(indicators, 1)
    countryFilter    = random.sample(countries, 2)
    
    # how many rows are there that have this country name, this indicator, and this are between the specified years?. 
    # Mesh returns a bool vector which we apply on to our data
    filterMesh = (data['CountryName'] == countryFilter[0]) & (data['IndicatorName'].isin(indicatorsFilter)) & \
                    (data['Year'].isin(yearsFilter))
    
    # which rows have this condition to be true? - We apply the mesh filter to find out
    filteredData1 = data.loc[filterMesh]
    # we then return only the columns we're interested in
    filteredData1 = filteredData1[['CountryName','IndicatorName','Year','Value']]

    # %s and %d are used for **string formatting** -> '%s' acts a placeholder for a string while '%d' acts as a placeholder
    # for a number (decimal). Their associated values are passed in via a tuple using the % operator
    
    # we need to print out the below only when our while condition is True
    if(len(filteredData1) < len(yearsFilter)-1):
        print('Skipping ... %s since very few rows (%d) were found' % (indicatorsFilter, len(filteredData1)))
```

```python
# What did we pick eventually? 
indicatorsFilter
```

```python
# the len filterData1 lets us know how many rows of data were retrieved

len(filteredData1)
```

```python
'''
Country 2
'''

while(len(filteredData2) < len(filteredData1)-1):
    filterMesh = (data['CountryName'] == countryFilter[1]) & (data['IndicatorName'].isin(indicatorsFilter)) & \
                (data['Year'].isin(yearsFilter))
    filteredData2 = data.loc[filterMesh]
    filteredData2 = filteredData2[['CountryName','IndicatorName','Year','Value']]
    old = countryFilter[1]
    # below picks a new country. Filtered data remains for the previous country if it meets the criteria
    countryFilter[1]    = random.sample(countries, 1)[0]
    
    if(len(filteredData2) < len(filteredData1)-1):
        print('Skipping ... %s, since very few rows (%d) found' % (old, len(filteredData2)))
```

```python
if len(filteredData1) < len(filteredData2):
    small = len(filteredData1)
else:
    small = len(filteredData2)
```

```python
filteredData1=filteredData1[0:small]
filteredData2=filteredData2[0:small]
```

## Example: Scatter Plot

Now that we have the data for two countries for the same indicators, let's plot them using a scatterplot.

```python
%matplotlib inline
import matplotlib.pyplot as plt

# as usual we use can the .subplots method to return a tuple with the figure, axis attributes. doing this, we can then use
# methods on the axis as below
fig, axis = plt.subplots()

# Grid lines, Xticks, Xlabel, Ylabel

axis.yaxis.grid(True)
# 'pad' parameter in the set_title() method allows you to create space between the plot title and the plot itself
axis.set_title(indicatorsFilter[0],fontsize=11, pad=20)
axis.set_xlabel(filteredData1['CountryName'].iloc[0],fontsize=10)
axis.set_ylabel(filteredData2['CountryName'].iloc[0],fontsize=10)

# below returns a pandas Series for Values of the filtered data for countries 1 and 2
X = filteredData1['Value']
Y = filteredData2['Value']

# we then plot the series as a scatter plot using the matplotlib.pyplot scatter() method. remember the scatter plot presents
# the data for year incrementally, but doesn't label year specifically. 
axis.scatter(X, Y)
plt.show()
```

## Example: Line Plot

Here we'll plot the indicator over time for each country. First country 1:

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14, 8))

# .set_ylim() method sets the y-axis view limits (lower, upper)
ax.set_ylim(min(0,filteredData1['Value'].min()), 2*filteredData1['Value'].max())
ax.set_title('Indicator Name : ' + indicatorsFilter[0])

# .plot() is used to plot line plots
# 'r--' parameter specifies a red dotted line
# label argument specifies the label for the data - in below it's the country 1 name
ax.plot(filteredData1['Year'], filteredData1['Value'], 'r--', label=filteredData1['CountryName'].unique()[0]) 

# the .legend() method adds a legend. has various parameters including loc, shadow, and prop (font properties) among others
legend = plt.legend(loc = 'upper center', 
                    shadow=False,
                    prop={'weight':'roman','size':'xx-large'})

# a rectangle can be added around the legend by assigning it to a new variable and using the .get_frame() method
# this turns it in to a matplotlib.patches.FancyBboxPatch object, upon which desired methods can be used
frame = legend.get_frame()
print(type(frame))
# set_facecolor() is a method that can be applied on FancyBbxPatch objects to change the shade of the interior
frame.set_facecolor('.9')
plt.show()
```

And for country 2:

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(14, 8))

# below adjusts the lower and upper limit with the aim to bring the graph to center
ax.set_ylim(min(0,filteredData2['Value'].min()), 2*filteredData2['Value'].max())

ax.set_title('Indicator Name : ' + indicatorsFilter[0])
# this time we're specifying data for country 2 in the line plot x, y data arguments
ax.plot(filteredData2['Year'], filteredData2['Value'] ,
         label=filteredData2['CountryName'].unique()[0],
         # color, lw (line width) and ls (linestyle) specify parameters of the line
         color="purple", lw=1, ls='-', 
         # marker parameter specifies the marker style for each data point
         marker='s', markersize=15, 
         # markers can be further specified with below showing some of the possibilities
         markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="blue") 

# adding a legend
legend = plt.legend(loc = 'upper right', 
                    shadow=True,
                    prop={'weight':'roman','size':'xx-large'})

# Similar to the country 1 plot, we can add a rectangle around the legend using the get_frame() method and specifying 
# parameters
frame = legend.get_frame()
frame.set_facecolor('.9')
plt.show()
```

# Example (random datasets)

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

countof_angles = 36
countof_radii  = 8

# numpy.linspace() returns evenly spaced numbers over a specified interval.
# paramteres are numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]

# array - radii
array_rad = np.linspace(0.125, 1.0, countof_radii)

# array - angles
array_ang = np.linspace(0, 2*np.pi, countof_angles, endpoint=False)

# numpy.repeat() repeats elements of an array. Arguments are numpy.repeat(a (input array), # of repeats, axis=None)[source]
# newaxis() is used to increase the dimension of the existing array by one more dimension

# repeat all angles per radius
array_ang = np.repeat(array_ang[...,np.newaxis], countof_radii, axis=1)

# from polar (radii, angles) coords to cartesian (x, y) coords
x = np.append(0, (array_rad*np.cos(array_ang)).flatten())
y = np.append(0, (array_rad*np.sin(array_ang)).flatten())

# saddle shaped surface
z = np.sin(-x*y)

fig = plt.figure(figsize=(20,10))
ax  = fig.gca(projection='3d')

# .plot_trisurf() method can be used to plot 3d charts
ax.plot_trisurf(x, y, z, cmap=cm.autumn, linewidth=.25)

plt.show()
# .savefig method is used to save a figure as an img, e.g. .png file
fig.savefig("vis_3d.png")
```

# Example (random dataset)

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

n_points = 200

# random.rand creates an array of the given shape and populates it with random samples from a uniform distribution over
# [0, 1). if no shape given, will return a rank 1 array

radius   = 2 * np.random.rand(n_points)
print(radius.shape)
angles   = 2 * (np.pi) * np.random.rand(n_points)
area     = 400 * (radius**2) * np.random.rand(n_points)

colors = angles

fig = plt.figure(figsize=(20,10))
ax  = plt.subplot(111, polar=True)

c = plt.scatter(angles, radius, c=colors, s=area, cmap=plt.cm.hsv)
c.set_alpha(1.95)

plt.show()
fig.savefig("vis_bubbleplot.png")
```

# Example 4: Box Plots (random datasets)

```python
# numpy.random.seed seeds the generator.

np.random.seed(452)

# Three ararys of 100 points each
A1 = np.random.normal(0, 1, 100)
A2 = np.random.normal(0, 2, 100)
A3 = np.random.normal(0, 1.5, 100)

# Concatenate the three arrays
data = [ A1, A2, A3 ]

# setting nrows and ncols determines how many subplot figs and axes are returned. here, we return 2 graphs
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))


# Box plot: Notch Shape
bplot1 = axes[1].boxplot(data,
                         notch=True,  
                         vert=True,           # vertical aligmnent
                         patch_artist=True)   # color

# Box plot: Rectangular 
bplot2 = axes[0].boxplot(data,
                         vert=True,           # vertical aligmnent
                         patch_artist=True)   # color




colors = ['tomato', 'darkorchid', 'lime']
# more colors here: http://matplotlib.org/examples/color/named_colors.html

for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# Grid lines, Xticks, Xlabel, Ylabel
for axis in axes:
    axis.yaxis.grid(True)
    axis.set_xticks([y for y in range(len(data))], )
    axis.set_xlabel('Sample X-Label',fontsize=20)
    axis.set_ylabel('Sample Y-Label',fontsize=20)

    
# Xtick labels
plt.setp(axes, xticks=[y for y in range(len(data))],
         xticklabels=['X1', 'X2', 'X3'])

plt.show()
fig.savefig("vis_boxplot.png")
```
